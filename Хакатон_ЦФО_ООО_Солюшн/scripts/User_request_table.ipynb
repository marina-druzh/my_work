```python
import pandas as pd
import gradio as gr

# create a sample DataFrame
df = pd.read_csv('work_directory/merged_df.csv')

# get the unique values from the "category" column
category_list = df['Регион'].unique().tolist()

# define your function to get data based on the selected category
def get_data(region):
    data = df[df['Регион'] == region]
    return data

# create the Gradio dropdown using the unique values
dropdown = gr.inputs.Dropdown(choices=category_list)

# use the dropdown in your Gradio interface
interface = gr.Interface(fn=get_data, inputs=dropdown, outputs=["dataframe"])
interface.launch()
```
