```python
# Определение абзацев для каждого растения
def make_plant_dict(text):
    
    pl_dict = {}
    for i in abstracts:
        pos = text.find(i)
        if pos != -1:
            pl_dict[pos] = i
        else:
            continue
    sorted_plant_dict = sorted(pl_dict.items(), key=lambda value: value[0])
    return sorted_plant_dict 
```

```python
# Создание окончателной таблицы растений с колонками по абзацам 
def make_table(plant_data):
    keys = plant_name
    plant_dict_all = []

    for i, key in enumerate(keys):
        abstracts = make_plant_dict(plant_data[key][0])
#         abstracts = plant_dict[key]
        details = {}
        for j in range(len(abstracts)-1):
            row = ''.join(plant_data[key]).strip()

            start_pos = row.find(abstracts[j][1])
            end_pos = row.find(abstracts[j+1][1])
            details['Название.'] = key
            details[abstracts[j][1]] = row[start_pos+len(abstracts[j][1]):end_pos].strip()
            if j == (len(abstracts) - 2):
                start_pos = row.find(abstracts[j+1][1])
                details[abstracts[j+1][1]] = row[start_pos+len(abstracts[j+1][1]):].strip()

        plant_dict_all.append(details)
    return plant_dict_all
```

```python
abstracts = ['Описание.', 'Ареал.', 'Экология.', 'Ресурсы.', 'Возделывание.', "Выращивание.", 'Химический состав.', 
             'Сырье.', 'Фармакологические свойства.', 'Применение в медицине.']
```

```python
plant_info_all = make_table(plant_data)
```
